```python
from simulation import Simulation
import matplotlib.pyplot as plt
import numpy as np
from functions import *
import pandas as pd
```

```python
simulation = Simulation(Np=2000,initial_velocity=100)
exe = 0
tempo = 100 #inserir um input depois
#print(f'execução número {exe}')
while tempo > simulation.time:
    print('---------------------')
    exe += 1
    print(f'execução número {exe}')
    simulation.mainloop()
#print(f'as pressões na simulação foram {simulation.pressures}')
#print(f'Os tempos das colisões na simulação foram {simulation.time_wall_collision}')
print(f'o desvio padrão da pressão {np.std(simulation.pressures)}')
```

```python

scalar_velocities = np.array([])
for particle in simulation.particles:
    scalar_velocities = np.hstack((scalar_velocities,magnitude(particle.velocity)))
scalar_velocities = np.sort(scalar_velocities)
scalar_velocities
```

```python
velocities = np.zeros((1,2))
for particle in simulation.particles:
    velocities = np.vstack((velocities,particle.velocity))

velocities[1]
```
